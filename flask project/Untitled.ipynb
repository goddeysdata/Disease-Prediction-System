```python
### importing libraries  
import pickle
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier

# Load dataset and preprocess
disease_df = pd.read_csv("dataset.csv")
disease_df.fillna("None", inplace=True)

# Encode target variable
encoder = LabelEncoder()
disease_df["Disease"] = encoder.fit_transform(disease_df["Disease"])

# Encode symptoms
symptom_cols = disease_df.columns[1:]
symptom_encoder = LabelEncoder()
for col in symptom_cols:
    disease_df[col] = symptom_encoder.fit_transform(disease_df[col])

# Train Decision Tree model
X = disease_df[symptom_cols]
y = disease_df["Disease"]
model = DecisionTreeClassifier(random_state=42)
model.fit(X, y)

# Save model and encoders
with open("decision_tree_model.pkl", "wb") as f:
    pickle.dump(model, f)

with open("label_encoder.pkl", "wb") as f:
    pickle.dump(encoder, f)

with open("symptom_encoder.pkl", "wb") as f:
    pickle.dump(symptom_encoder, f)

print("Model and encoders saved successfully!")
```
